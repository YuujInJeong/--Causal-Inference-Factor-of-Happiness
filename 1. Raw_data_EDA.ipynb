```python
import pandas as pd

# Load your DataFrame

def set_binary_flag(row, columns):
    for col in columns:
        new_col_name = col + '_여부'  # Append '_여부' to each column name to create the new column names
        row[new_col_name] = 1 if row[col] == 1 else 0
    return row

def process_expense_categories(row):
    expense_dict = {
        'h15cr001': '기부금',
        'h15cc003': '외식비',
        'h15cd001': '주류비',
        'h15ce001': '담배구입비',
        'h15ch001': '문화생활비_서적등',
        'h15ch003': '문화생활비_박물관등',
        'h15ch005': '문화생활비_경마권',
        'h15ch010': '신문_정기구독료',
        'h15ck007': '스포츠레저_장비구입비',
        'h15ck003': '악기구입비',
        'h15cn006': '국내여행_숙박료',
        'h15cn009': '국외여행_숙박료'
    }
    for col, new_name in expense_dict.items():
        row[new_name + '_지출여부'] = 1 if row[col] == 1 else 0
    return row

# Applying the functions to the DataFrame
columns_to_process = [
    'h15fb009', 'h15fb011', 'h15cq005', 'h15cr001', 'h15cc003', 'h15cd001', 'h15ce001',
    'h15ch001', 'h15ch003', 'h15ch005', 'h15ch010', 'h15ck007', 'h15ck003', 'h15cn006', 'h15cn009'
]
df_final2 = df5.apply(lambda row: set_binary_flag(row, columns_to_process), axis=1)
df_final2 = df5.apply(process_expense_categories, axis=1)
```

```python
import pandas as pd

# 문화생활비 관련 항목을 종합하여 하나라도 1이면 1을 설정하는 함수
def process_cultural_activities(row):
    # '문화생활비_서적등', '문화생활비_박물관등', '문화생활비_경마권' 열 중 하나라도 1이면 '문화생활비_종합'을 1로 설정
    if row['문화생활비_서적등_지출여부'] == 1 or row['문화생활비_박물관등_지출여부'] == 1 or row['문화생활비_경마권_지출여부'] == 1:
        row['문화생활비_종합'] = 1
    else:
        row['문화생활비_종합'] = 0
    return row

# 기존의 DataFrame df5에 대해 set_binary_flag 및 process_expense_categories 함수 적용
df_final2 = df5.apply(lambda row: set_binary_flag(row, columns_to_process), axis=1)
df_final2 = df_final2.apply(process_expense_categories, axis=1)
df_final2 = df_final2.apply(process_cultural_activities, axis=1)
```
