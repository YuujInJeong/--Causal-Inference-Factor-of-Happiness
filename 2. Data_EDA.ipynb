```python
import pandas as pd 
import numpy as np
pd.set_option("display.max_columns", 500)
```

```python
df = pd.read_csv('causal_inference_personal_data.csv', index_col=0)
```

```python
df
```

#### Nan -> -1

```python
df = df.fillna(-1)
```

#### (-) 값을 갖는 컬럼별 개수 찾기

```python
pd.options.display.max_rows = len((df < 0).sum().sort_values())


# 결과 출력
print((df < 0).sum().sort_values())
```

#### Null값 개수를 기준으로 drop

```python
df_drop_14000 = df.drop(df.columns[(df < 0).sum() > 14000], axis=1)
df_drop_14000.shape
```

```python
df_drop_10000 = df.drop(df.columns[(df < 0).sum() > 10000], axis=1)
df_drop_10000.shape
```

```python
df_drop_10000.head()
```

```python
pid_hid_col = [
    x for x in df_drop_10000.columns 
    if ("h15" in x) or 
        ("pid" in x) or 
        ("hid" in x) or 
        ("ps" in x) or 
        ("inc" in x) or 
        ("gh" in x) or
        ("wt" in x) or 
        ("gj" in x)]


df_final_1 = df_drop_10000.drop(pid_hid_col, 
                                axis=1)

df_final_1 = df_final_1.drop([
                "p15aa102",
                'p15aa104',
                 'p15aa105',
                 'p15aa106',
                 'p15aa107',
                 'p15aa108',
                 'p15aa109',
                 'p15aa110',
                 'p15aa008',
                 'p15aa009',
                 'p15bc014',
                 'version15'
                 ], axis=1)


df_final_1 = df_final_1.applymap(lambda x: max(0, x))
```

#### 고유한 값이 20개 이상인 경우 10분위 quantile로 변환함.

```python
import pandas as pd

# 새로운 데이터프레임을 만듭니다.
quan_df = df_final_1.copy()

# 데이터프레임의 각 열에 대해 반복합니다.
for col in df_final_1.columns: 
    # 열의 고유값 개수가 100개 이상인 경우에만 처리합니다.
    if len(df_final_1[col].unique()) >= 20:
        # 중복된 경계를 제거하고 데이터를 10개의 구간으로 나눕니다.
        quan_df[col] = pd.qcut(df_final_1[col], q=10, labels=False, duplicates="drop")
        # 구간의 레이블을 1부터 10까지로 변경합니다.
        quan_df[col] += 1
```

```python
for col in quan_df.columns: 
    print(col, len(quan_df[col].unique()))
```
